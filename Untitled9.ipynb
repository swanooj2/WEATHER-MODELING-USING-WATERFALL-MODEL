```python
 import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic(x, a=1, b=-2, c=3):  # Adjusted the coefficients
    return a * x**2 + b * x + c  # y = ax^2 + bx + c

# Define the stages of the Waterfall Model
stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]

# Equally spaced x positions for each stage
x_positions = np.linspace(-2, 2, len(stages))  # Changed the range of x values
y_positions = quadratic(x_positions)  # Compute y positions using the quadratic function

# Create the quadratic curve for visualization
x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(x_curve, y_curve, color="purple", linestyle="--", label="Quadratic Curve")  # Changed line color

# Add markers and connect them with vertical lines to simulate the waterfall effect
for i in range(len(stages)):
    if i > 0:
        # Draw a horizontal line (waterfall effect)
        plt.plot([x_positions[i-1], x_positions[i]], [y_positions[i-1], y_positions[i-1]], color="orange", linewidth=2)
        # Draw a vertical line (waterfall effect)
        plt.plot([x_positions[i], x_positions[i]], [y_positions[i-1], y_positions[i]], color="orange", linewidth=2)

    # Place markers and labels
    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5)
    plt.text(x_positions[i], y_positions[i] + 0.3, stages[i], ha="center", fontsize=12, color="black")

# Style the plot
plt.title("Waterfall Model Representation Using Quadratic Curve", fontsize=16)
plt.xlabel("Stages", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
